```python
tl = pd.read_csv('train.csv')
```

```python
al = len(tl)
bl = len(tl['beer/beerId'].unique())
ul = len(tl['user/profileName'].unique())
print('Total Reviews: {a}\nTotal Beers: {b}\nTotal Users: {d}\nReviews/Beer: {c}'.format(a=al, b=bl, c=round((al/bl), 2), d=ul))
```

```python
tl_pivot = (tl.pivot_table(index='beer/beerId', columns='user/profileName', values='review/overall')).fillna(0)
tl_pivot.head()
```

```python
beer_rating_count = (tl.
                     groupby(by=['beer/beerId'])['review/overall'].
                     count().
                     reset_index().
                     rename(columns={'review/overall':'beer_ratingcount'})
                    [['beer/beerId', 'beer_ratingcount']])
print(beer_rating_count['beer_ratingcount'].describe())
```

```python
user_rating_count = (tl.
                     groupby(by=['user/profileName'])['review/overall'].
                     count().
                     reset_index().
                     rename(columns={'review/overall':'usr_ratingcount'})
                    [['user/profileName', 'usr_ratingcount']])
print(user_rating_count['usr_ratingcount'].describe())
```

```python
tl_with_beercount = tl.merge(beer_rating_count, left_on = 'beer/beerId', right_on = 'beer/beerId')
tl_with_beercount.head(1)
```

```python
tl_counts = tl_with_beercount.merge(user_rating_count, left_on = 'user/profileName', right_on = 'user/profileName')
tl_counts.head(1)
```

```python
beer_threshold = 5
usr_threshold = 5
popular_beers = tl_counts.query('beer_ratingcount >= @beer_threshold')
popular_beers = popular_beers.query('usr_ratingcount >= @usr_threshold')
popular_beers = popular_beers.reset_index(drop=True)
popular_beers.head(1)
```

```python
print('Unique Beers:', len(popular_beers['beer/beerId'].unique()),'\nUnique Reviewers:', len(popular_beers['user/profileName'].unique()))
```
